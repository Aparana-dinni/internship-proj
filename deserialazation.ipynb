```python
import numpy as np
from pickle import load
```

```python
# Loading pretrained classifiers from pickle file
from google.colab import drive
drive.mount('/content/drive')

import os
os.chdir('/content/drive/MyDrive')

scaler = load(open('/content/drive/MyDrive/model/standard_scaler.pkl', 'rb'))
knn = load(open('./model/knn_model.pkl', 'rb'))
regressor = load(open('./model/lr_model.pkl', 'rb'))
rf = load(open('./model/rf_model.pkl', 'rb'))
xgb_reg = load(open('./model/xgb_model.pkl', 'rb'))
```

```python
# Read the Query Data

print("Enter ablong shell Details")
Sex = int(input('Enter the sex: '))
Length = float(input('Enter the SepalLengthWidth (in cm): '))
Diameter = float(input('Enter the Diameter (in cm): '))
Height = float(input('Enter the Height (in cm): '))
Whole_weight = float(input('Enter the Whole_weight (in cm): '))
Shucked_weight = float(input('Enter the Shucked_weight (in cm): '))
Viscera_weight = float(input('Enter the Viscera_weight (in cm): '))
Shell_weight =  float(input('Enter the Shell_weight (in cm): '))
```

```python
query_point = np.array([Sex, Length, Diameter, Height, Whole_weight,Shucked_weight,Viscera_weight, Shell_weight])
```
